<h3 style="color: mediumseagreen" id="probabilidad">Probabilidad</h3>

游 [Regresar al 칈ndice](#INDICE)

Medida de la certidumbre de que ocurra un evento. Su valor es un n칰mero entre 0 y 1, donde un evento imposible corresponde a `cero` y un evento seguro corresponde a `uno`

* La teor칤a de la probabilidad tiene como `objetivo` definir una estructura matem치tica que describa salidas aleatorias de experimentos.
    * Experimento aleatorio: experimentos en los que no se puede predecir su resultado.
        * ejemplo: Cuando arrojamos una moneda, no podemos predecir con certeza el resultado individual, pero al repetir el experimento muchas veces, podemos observar una tendencia en el promedio de los resultados.

* La probabilidad moderna esta basada en un conjunto de axiomas propuestos por `Kolmogorov`, que introduce tres conceptos:
    * Espacio muestral $(\Omega)$: En probabilidad es el conjunto de todos los resultados posibles de un experimento aleatorio.
    * Evento $(A)$: Uno o varios de los posibles resultados del experimento.
    * Probabilidad $(p)$: grado de creencia de que un evento ocurra.
    * `Ejemplo 1:`
    
        Sea $\varepsilon$(Experimento aleatorio) = sale un n칰mero par de lanzar un dado.
        1. $\Omega={1,2,3,4,5,6}$ 
        2. $A={2,4,6}$
        3. $p(A)=\frac{numero de elementos del conjunta A}{cantiad de elementos de \Omega}$
    * `Ejemplo 2:`
    
        Sea $\varepsilon$ = sacar dos monedas de una maleta.
        *Contexto: la maleta contiene monedas de USA(\\$) y UK(춲) Se asume que la probabilidad de sacar aleatoriamente un `\\$ es 0.3`*
        * Se tiene un $\Omega$=(\\$,\\$),(\\$,춲),(춲,\\$),(춲,춲)
        * Nuestro evento viene relacionado por la cantidad de veces que obtenemos un \$.
            * (\\$,\\$) = 2
            * (\\$,춲) = 1
            * (춲,\\$) = 1
            * (춲,춲) = 0
            
        Para referirnos a las probabilidades sobre las cantidades de interes se usa la letra $\tau$ donde para el espacio muestral:
            * $\tau={0,1,2}$
        
        * Finalmente, la probabilidad para cada evento en cuesti칩n.
        
        **P(X=2)**
                $$P(X=2) = P(\$,\$)    
                =P(\$)\cdot P(\$)  
                =0.3\cdot0.3=0.09$$
                
        **P(X=1)**
                $$P(X=2) = (P(\$,춲)\cup P(춲,\$))     
                =P(\$,춲) + P(춲,\$) 
                =0.3\cdot (1-0.3) + (1-0.3)\cdot 0.3=0.42$$
        
        **P(X=0)**
                $$P(X=2) = P(춲,춲)    
                =P(춲)\cdot P(춲)  
                =(1-0.3)\cdot (1-0.3)=0.49$$

La probabilidad marginal de cada variable aleatoria, puede ser vista como la suma sobre una fila o columna

$$P(X=x_{i})=\frac{c_{i}}{N}=\frac{\sum_{j=1}^{3}n_{ij}}{N}$$

$$P(Y=y_{i})=\frac{r_{i}}{N}=\frac{\sum_{j=1}^{5}n_{ij}}{N}$$

donde se asume que ambas sumatorias son iguales a `1`. $n_{ij}=n칰mero de eventos del estado$

Las probabilidades condicionales son una fracci칩n de una fila o columna en una celda en particular:

$$P(Y = y_{j} | X = x_{i}) = \frac{n_{ij}}{c_{i}}$$

$$P(X = x_{i} | Y = y_{j}) = \frac{n_{ij}}{r_{j}}$$

* $c_{i}$ = suma de las frecuencias individuales para cada $ith$ columna
* $r_{j}$ = suma de las frecuencias individuales para cada $jth$ fila

<h4 style="color: seagreen" id="prob-continua">Probabilidad Continua</h4>

游 [Regresar al 칈ndice](#INDICE)

En esta secci칩n consideramos variables aleatorias con valores reales, es decir, consideramos espacios objetivo que son intervalos de la l칤nea real $\Re$

En espacios continuos, existen dos particularidades adicionales que pueden resultar contraintuitivas. En primer lugar, el conjunto de todos los subconjuntos (utilizados para definir el espacio de eventos $A$) no es lo suficientemente manejable. $A$ debe estar restringido para comportarse adecuadamente en cuanto a complementos de conjuntos, intersecciones y uniones de conjuntos. En segundo lugar, el tama침o de un conjunto (que en espacios discretos se puede obtener contando los elementos) resulta ser complicado. El tama침o de un conjunto se llama su medida. Por ejemplo, la cardinalidad de conjuntos discretos, la longitud de un intervalo en $\Re$ y el volumen de una regi칩n en $\Re^{d}$ son todas medidas. Los conjuntos que se comportan bien bajo operaciones de conjunto y adem치s tienen una topolog칤a se llaman 픢-치lgebra de Borel.

**Funci칩n de densidad de probabilidad: f(x)** nos ayuda a calcular la probabilidad de que la variable aleatoria X asuma un valor en el intervalor que va desde a hasta b
    * X: variable aleatoria continua (incontable)
    * f(x): funci칩n de densidad de probabilidad
    
$$\int_{a}^{b}f(x) \,dx$$

Condiciones:
1. $f(X)>=0$
2. area bajo la curva de densidad es igual a 1

`Ejemplo`
La variable aleatoria continua X tiene la siguiente funci칩n de densidad de probabilidad
$$f(x) = \begin{cases}
0.25, & \text{si } 0 \leq x \leq 4 \\
0, & \text{de lo contrario}
\end{cases}$$

Calcular:
1. $P(1<=x<=3)$
2. $P(x=1)$
3. $P(1<x<=3)$
4. $P(1<x<=4)$

```python
# Soluci칩n
import numpy as np
import matplotlib.pyplot as plt

# Definir los puntos de x y los valores correspondientes de f(x)
x = np.linspace(-2, 6, 1000)
y = np.piecewise(x, [x < 0, (0 <= x) & (x <= 4), x > 4], [0, 0.25, 0])

# Crear la gr치fica
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Ejercicio')
plt.axvline(x=0,color='r',linestyle='--',label='x=0')
plt.axvline(x=1,color='orange',linestyle='--',label='x=1')
plt.axvline(x=3,color='g',linestyle='--',label='x=3')
plt.axvline(x=4,color='purple',linestyle='--',label='x=4')
plt.legend()
plt.grid(True)

print(f"1. Probabilidad = {(3-1)*(0.25)*100:.2f}%")
print(f"2. Probabilidad = {0} No admite valores discretos")
print(f"3. Probabilidad = {(3-1)*(0.25)*100:.2f}% P(a<x<=b)=P(a<=x<=b) ya que p(a=x) = 0") 
print(f"4. Probabilidad = {(4-1)*(0.25)*100:.2f}%")
```

**Funci칩n de distribuci칩n acumulativa: F(x)** Es la funci칩n que para un valor x, nos da la probabilidad de que la variable aleatoria sea menor o igual que dicho valor x, es decir:

$$F(x) = P(X<=x)$$
* Variable aleatoria = X (may칰scula)
* valor de la variable aleatoria = x (min칰scula)

`Ejemplo:`
$$F(x) = \begin{cases}
0, & \text{si } x < 0 \\
0.2x, & \text{si } 0 \leq x \leq 5 \\
1, & \text{si } x > 5
\end{cases}$$

1. P(X<=2)
2. P(X<=3)
3. P(X>2)
4. P(2<=X<=3)

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir un rango de valores de x
x = np.linspace(-1, 10, 400)  # Puedes ajustar el rango seg칰n tus necesidades

# Definir la funci칩n F(x)
def F(x):
    if x < 0:
        return 0
    elif 0 <= x <= 5:
        return 0.2 * x
    else:
        return 1

# Aplicar la funci칩n F a cada valor de x
y = [F(xi) for xi in x]

# Crear la gr치fica
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(x, y, label='$F(x)$')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('$F(x)$')
plt.title('Gr치fica de $F(x)$')
plt.axvline(x=3,color='purple',linestyle='--',label='x=3')
plt.axvline(x=2,color='r',linestyle='--',label='x=2')
plt.legend()

# Colorear el 치rea
plt.fill_between(x, y, where=(x >= 2) & (x <= 3), color='crimson', alpha=0.5)
plt.fill_between(x, y, where=(x < 2), color='gold', alpha=0.5)
plt.fill_between(x, y, where=(x > 3), color='lime', alpha=0.5)

# Mostrar la gr치fica
plt.grid(True)
plt.show()
```

```python
cuarto_punto = segundo_punto - primer_punto
print(f"1. Su probabilidad es: {primer_punto*100:.2f}%")
print(f"2. Su probabilidad es: {segundo_punto*100:.2f}%")
print(f"3. Su probabilidad es: {tercer_punto*100:.2f}%")
print(f"4. Su probabilidad es: {cuarto_punto*100:.2f}%")
```

1. Distribuci칩n binomial

    Esta distribuci칩n se aplica cuando la probabilidad del 칠xito esperado siempre ser치 la misma.
    `Por ejemplo: `
    * Al arrojar una moneda n veces y salga cara x veces.
    * Al arrojar un dado 10 veces y salga 4 veces la cara 1.
    * Dentro de un grupo de 10 personas la probabilidad de que contraigan alguna enfermedad 5 personas.
    
    Esta compuesta por el tama침o de la muestra, n칰mero de 칠xitos esperados y la probabilidad de 칠xito.
    
    $$P(x,n,p) = \begin{pmatrix} n\\x \end{pmatrix} \cdot p^x \cdot (1-p)^{n-x}$$

```python
import math

def binomial_probability(x, n, p):
    # Calcula la combinaci칩n binomial utilizando math.comb
    binomial_coefficient = math.comb(n, x)
    
    # Calcula la probabilidad usando la f칩rmula
    probability = binomial_coefficient * (p**x) * ((1-p)**(n-x))
    
    return probability

n = 10  # N칰mero de lanzamientos del dado
x = 4   # N칰mero de veces que se obtiene el n칰mero 1
p = 1/6  # Probabilidad de obtener el n칰mero 1 en un lanzamiento

probability = binomial_probability(x, n, p)
print(f"La probabilidad de obtener el n칰mero 1 exactamente 4 veces al lanzar un dado 10 veces es: {probability*100:.2f}%")
```

2. Distribuci칩n Hipergeom칠trica

    Esta distribuci칩n normalmente se plantean que de una poblaci칩n o conjunto "N", se extrae una muestra "n", dentro de la cual se espera que "x" veces ocurra un evento que tambi칠n esta ocurriendo "k" veces en la poblaci칩n.
    `Por ejemplo: `
    * Poblaci칩n de vacas, 6 con manchas negras y 4 con manchas cafes => N
    * Muestra => n => 3 vacas
    * Evento => x => se esperan 2 vacas con manchas cafes.
    * En la poblaci칩n identificamos las vacas con manchas cafes => k => 4 vacas con manchas cafes.
    
    $$P(x,m,k,N) = \frac{\begin{pmatrix} k\\x\end{pmatrix}\begin{pmatrix}N-k\\n-x\end{pmatrix}}{\begin{pmatrix}N\\n\end{pmatrix}}$$

```python
from scipy.special import comb

def calculate_probability(x, m, k, N):
    numerator = comb(k, x) * comb(N - k, m - x)
    denominator = comb(N, m)
    probability = numerator / denominator
    return probability

x = 2  # Reemplaza con el valor deseado de x
m = 3  # Reemplaza con el valor deseado de m
k = 4  # Reemplaza con el valor deseado de k
N = 10  # Reemplaza con el valor deseado de N

result = calculate_probability(x, m, k, N)
print(round(result*100,2),"%")
```

3. Distribuci칩n de Poisson

    * Habla de promedios de ocurrencias dentro de un periodo determinado $\lambda$
    * Plantea el experimento en una poblaci칩n grande y con una probabilidad de 칠xito peque침a $\lambda = N \cdot p$, donde N es el tama침o de la poblaci칩n y p la probabilidad de 칠xito
    
$$P(x,\lambda) = \frac{e^{-\lambda}\cdot \lambda^x}{x!}$$

donde x es el **N칰mero de 칠xitos esperados**

`Ejemplo: `

```python
import math

def poisson_probability(x, lmbda):
    return (math.exp(-lmbda) * lmbda**x) / math.factorial(x)

x = 3  # Reemplaza con el valor deseado de x
lmbda = 2.5  # Reemplaza con el valor deseado de lambda

probability = poisson_probability(x, lmbda)
print("La probabilidad de obtener {} eventos en un intervalo con lambda={} es: {:.4f}".format(x, lmbda, probability))
```

Presenta un valor de m치xima frecuencia n, a partir del cual, decae hacia ambos lados con una simetr칤a perfecta. Esta simetr칤a hace que a valores situados a igual distancia del valor modal por izquierda y por derecha de la distribuci칩n, les corresponda la misma probabilidad.



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los par치metros de la distribuci칩n normal
mu = 0  # Media
sigma = 1  # Desviaci칩n est치ndar

# Generar datos para la distribuci칩n normal
data = np.random.normal(mu, sigma, 1000)

# Crear un histograma para visualizar la distribuci칩n
fig, ax = plt.subplots(figsize=(8,3))
ax.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Calcular la funci칩n de densidad de probabilidad (PDF) de la distribuci칩n normal
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Graficar la funci칩n de densidad de probabilidad
plt.plot(x, pdf, 'r-', lw=2)

plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci칩n Normal')
plt.grid(True)

# Mostrar el gr치fico
plt.show()
```

Se usa cuando queremos encontrar la probabilidad de que 2 o mas eventos ocurran a la vez

2. Eventos dependientes:
    * $P(A\cap B)=P(A)\cdot P(B|A)$

`Ejemplo: `

En una urna hay 5 esferas azules, 2 rojas y una verde. Si se sacan 2 esferas consecutivas al azar `sin reemplazo`, C치lcule la probabilidad de:

```python
esferas = {'azules':5, 'rojas':2, 'verdes':1}

numero_casos = sum(esferas.values())
azules = esferas['azules']
prob_azules = azules/numero_casos
prob_verde_primera_azul = esferas['verdes']/(numero_casos-1) # ya que 1 esfera azul fue seleccionada
print(f"1. La probabilidad de que la primera sea azul y la segunda verde es de {prob_azules*prob_verde_primera_azul*100:.2f}%")
prob_rojas = esferas['rojas']/numero_casos
prob_rojas_primera_roja = (esferas['rojas']-1)/(numero_casos-1) # 1 roja menos en la urna
print(f"2. La probabilidad de que ambas sean rojas es de {prob_rojas*prob_rojas_primera_roja*100:.2f}%")
```

<h3 style="color:mediumseagreen" id="teorema-bayes">Teorema de Bayes</h3>

游 [Regresar al 칈ndice](#INDICE)

Es una aplicaci칩n de la probabilidad condicional.

Con el teorema, si tenemos $P(A|B)$, podemos llegar a $P(B|A)$

Est치 hecho para aplicarse varias veces consecutivas. Se usa en:

1. Medicina
2. Tecnolog칤a
3. Neurolog칤a
4. Negocios

$P(A|B) = \frac{P(B|A)\cdot P(A)}{P(B)}$

Si los eventos $A_1,A_2,...,A_n$ son mutuamente excluyentes con $P(A_i)\neq 0.$.Si ocurre un evento B con $P(B)\neq 0$ Entonces

$$P(A_k|B) = \frac{P(B|A_k)\cdot P(A_k)}{\sum_{i=1}^{n} P(B|A_i)\cdot P(A_i)}$$

`EJEMPLO`

```python
# Variables
total_urnas = 2
total_bolas_urna = 5
bolas_azules_urna_1 = 3
bolas_azules_urna_2 = 4
bolas_rojas_urna_1 = 2
bolas_rojas_urna_2 = 1

# probabilidades
prob_urna = 1/total_urnas
prob_azul_urna_1 = bolas_azules_urna_1/total_bolas_urna
prob_azul_urna_2 = bolas_azules_urna_2/total_bolas_urna
prob_roja_urna_1 = bolas_rojas_urna_1/total_bolas_urna
prob_roja_urna_2 = bolas_rojas_urna_2/total_bolas_urna

# probabilidad de que salga roja dentro de la urna 1 por probabilidad urna 1
prob_roja_urna1_urna1 = prob_roja_urna_1 * prob_urna
prob_roja_urna2_urna2 = prob_roja_urna_2 * prob_urna

# probabilidad de que salga roja
prob_roja = prob_roja_urna1_urna1 + prob_roja_urna2_urna2

# probabilidad de que salga la urna 1 si ya se extrajo una bola roja
bayes_teorem = prob_roja_urna1_urna1/prob_roja
print(f"La probabilidad de que salga de la urna 1 la bola roja es: {bayes_teorem*100:.2f}%")
```

<h4 style="color:red">Nota</h4>

La media, varianza y desviaci칩n est치ndar vistas son cuando tenemos las probabilidades de los datos, por lo cual es distinto de la media usual. Para ello de manera normal estas se trabajan con las siguientes formulas

1. **Media**

La media es una medida de tendencia central que se define como la suma de todos los valores de un conjunto de datos, dividida por el n칰mero de valores.

$\overline{X} = \sum_{n=1}^{n} X_i/N$, donde N es el total de datos

2. **Varianza**

$\sigma = \sqrt{\sigma^2}$

<h3 style="color:mediumseagreen" id="covarianza">Covarianza</h3>

游 [Regresar al 칈ndice](#INDICE)

La covarianza es una medida de la relaci칩n lineal entre dos variables aleatorias. Se define como la media de los productos de las desviaciones de cada variable de su media.

* $Cov(X, Y) = \frac{\sum_{i=1}^n (X_i - \mu_X)(Y_i - \mu_Y)}{n}$

* $Cov(X, Y) = \frac{\sum_{i=1}^n X_i\cdot Y_i}{n} - \overline{X}\cdot \overline{Y}$

donde:

* $X_i$ es el valor i-칠simo de la variable aleatoria X
* $Y_i$es el valor i-칠simo de la variable aleatoria Y
* $\mu_X$es la media de la variable aleatoria X
* $\mu_X$es la media de la variable aleatoria Y
* n es el n칰mero de observaciones

Interpretaci칩n:

* **Covarianza positiva:** Las dos variables tienden a moverse en la misma direcci칩n. A medida que un valor aumenta, el otro valor tambi칠n aumenta.
* **Covarianza negativa:** Las dos variables tienden a moverse en direcciones opuestas. A medida que un valor aumenta, el otro valor disminuye.
* **Covarianza cero:** No hay relaci칩n lineal entre las dos variables.

`Ejemplo`

```python
x = [6,4,8,9,4,8,10,9,5,6]
y = [5,5,7,5,3,8,10,8,7,6]

xy = sum([var[0]*var[1] for var in list(zip(x,y))])
media_x = np.mean(x)
media_y = np.mean(y)
n = len(x) if len(y)==len(x) else print("error")

covarianza = (xy/n) - (media_x*media_y)
print("La covarianza es",round(covarianza,2))
```

La magnitud de la covarianza se puede comparar con un rango de valores que indica la fuerza de la relaci칩n. En general, se considera que una covarianza es:

* Debil: Si su valor absoluto es menor que 0.3.
* Moderada: Si su valor absoluto es entre 0.3 y 0.6.
* Fuerte: Si su valor absoluto es mayor que 0.6.

Sin embargo, es importante tener en cuenta que estos rangos son solo una gu칤a general. La magnitud de la covarianza tambi칠n depende de la escala de las variables. Por ejemplo, una covarianza de 0.3 podr칤a ser considerada fuerte si las variables est치n en escalas muy grandes.

<h3 style="color:mediumseagreen" id="corr">Correlaci칩n</h3>

游 [Regresar al 칈ndice](#INDICE)

La correlaci칩n es una medida de la relaci칩n lineal entre dos variables aleatorias. Se define como el coeficiente de correlaci칩n de Pearson, que es una medida de la fuerza y la direcci칩n de la relaci칩n entre dos variables.

$r = \frac{\sum_{i=1}^n (X_i - \mu_X)(Y_i - \mu_Y)}{\sqrt{\sum_{i=1}^n (X_i - \mu_X)^2} \sqrt{\sum_{i=1}^n (Y_i - \mu_Y)^2}}$

donde:

* r es el coeficiente de correlaci칩n de Pearson.
* $X_i$ es el valor i-칠simo de la variable aleatoria X.
* $Y_i$ es el valor i-칠simo de la variable aleatoria Y.
* $\mu_X$ es la media de la variable aleatoria X.
* $\mu_Y$ es la media de la variable aleatoria Y.
* n es el n칰mero de observaciones.

Interpretaci칩n:

* **Correlaci칩n positiva:** Las dos variables tienden a moverse en la misma direcci칩n. A medida que un valor aumenta, el otro valor tambi칠n aumenta.
* **Correlaci칩n negativa:** Las dos variables tienden a moverse en direcciones opuestas. A medida que un valor aumenta, el otro valor disminuye.
* **Correlaci칩n cero:** No hay relaci칩n lineal entre las dos variables.

`Ejemplo`

```python
# Datos de ejemplo
conjunto1 = np.array([1, 2, 3, 4, 5])
conjunto2 = np.array([5, 4, 3, 2, 1])

# Calcula la matriz de correlaci칩n
matriz_correlacion = np.corrcoef(conjunto1, conjunto2)

# La matriz de correlaci칩n tendr치 la correlaci칩n entre conjunto1 y conjunto2 en la posici칩n (0,1) y (1,0)
correlacion = matriz_correlacion[0, 1]

print(f"La correlaci칩n entre conjunto1 y conjunto2 es: {correlacion:.2f}")
```
